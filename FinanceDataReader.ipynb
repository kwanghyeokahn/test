### FinanceDataReader는 Pandas-datareader를 대체하기 보다  보완하기 위한 목적으로 만들어졌다. 
### 종목코드 
- 거래소별 전체 종목코드 : KRX(KOSPI, KOSDAQ, KONEX),NASDAQ, NYSE, AMEX, S&P 500

### 가격 데이터
- 해외주식 가격 데이터 : AAPL(애플), AMZN(아마존), GOOG(구글) 등 ...
- 국내주식 가격 데이터 : 005930(삼성전자), 091990(셀트리온헬스케어) 등 ...
- 각종 지수 : KS11(코스피지수), KQ11(코스닥지수),DJI(다우지수), IXIC(나스닥지수), US500(S&P 5000)
- 환율 데이터 : USD/KRX(원달러 환율), USD/EUR (달러당 유로화 환율), CNY/KRW (위엔화 원화 환율)
- 암호화폐 가격 : BTC/USD(비트코인 달려가격, Bitfinex), BTC/KRW (비트코인 원화 가격, 빗썸)

KOSPI 종가, 상한, 하한 그래프

```python
plt.figure(figsize=(20,15))
plt.plot(df.index[:200],df['Close'][:200],'-g',df.index[:200],df['High'][:200],'-r',df.index[:200],df['Low'][:200],'-b') 
plt.show()
```

MACD

```python
plt.figure(figsize=(20,15))
ma26 = df['Close'][:200].rolling(window=26).mean()
ma12 = df['Close'][:200].rolling(window=12).mean()
macd_line = ma12 - ma26
ma9 = df['Close'][:200].rolling(window=9).mean()
signal_line = macd_line.rolling(window=26).mean()
plt.plot(df.index[:200],ma26,'-r', df.index[:200],ma12,'-b', df.index[:200],macd_line,'-g',  df.index[:200],signal_line,'-p')
```
